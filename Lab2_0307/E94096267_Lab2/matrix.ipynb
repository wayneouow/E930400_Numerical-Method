```python
import random
from copy import deepcopy

class Matrix:

    def __init__(self, nrows, ncols):
        """Construct a (nrows X ncols) matrix"""
        # create objects
        self.nrows = nrows
        self.ncols = ncols
        # create random nrows X ncols matrix
        # implement with 2-d list
        lists = [[] for k in range(nrows)]
        for i in range(nrows):
            for j in range(ncols):
                lists[i].append(random.randint(0,10))
        self.matrix = lists

    def add(self, m):
        """return a new Matrix object after summation"""
        # Check two matrix are in the same size
        if(self.nrows != m.nrows or self.ncols != m.ncols):
            print("Matrixs' size should be in the same size")
            return None
        # add self.matrix with m.matrix
        elif(self.nrows == m.nrows and self.ncols == m.ncols):
            #operating without modifying original matrix
            result = deepcopy(self)
            #scan & add
            for i in range(self.nrows):
                for j in range(self.ncols):
                    result.matrix[i][j] = self.matrix[i][j] + m.matrix[i][j]

        return result

    def sub(self, m):
        """return a new Matrix object after substraction"""
        # Check two matrix are in the same size
        if(self.nrows != m.nrows or self.ncols != m.ncols):
            print("Matrixs' size should be in the same size")
            return None
        # substraction self.matrix from m.matrix
        elif(self.nrows == m.nrows and self.ncols == m.ncols):
            # operating without modifying original matrix
            result = deepcopy(self)
            # scan & sub
            for i in range(self.nrows):
                for j in range(self.ncols):
                    result.matrix[i][j] = self.matrix[i][j] - m.matrix[i][j]

        return result

    def mul(self, m):
        """return a new Matrix object after multiplication"""
        # Check if the nrows of m is the same as ncols of self
        if(self.ncols != m.nrows):
            print("Cannot multiply two matrixs with size (", self.nrows, ",", self.ncols, ") and (", m.nrows, ",", m.ncols, ")")
            return None
        # Multiply self.matrix and m.matrix
        else:
            # create a new matrix with proper size
            result = Matrix(self.nrows, m.ncols)
            for i in range(self.nrows):
                for j in range(m.ncols):
                    # init new matrix
                    result.matrix[i][j] = 0;
                    # mul & sum
                    for k in range(self.ncols):
                        result.matrix[i][j] = result.matrix[i][j] + self.matrix[i][k] * m.matrix[k][j]
    
        return result

    def transpose(self):
        """return a new Matrix object after transpose"""
        # Tranpose
        # create proper size matrix
        result = Matrix(self.ncols, self.nrows)
        # swap rows & cols
        for i in range(self.nrows):
            for j in range(self.ncols):
                result.matrix[j][i] = self.matrix[i][j]

        return result
    
    def display(self):
        """Display the content in the matrix"""
        # print elements with matrix form
        for x in range(self.nrows):
            for y in range(self.ncols):
                print(self.matrix[x][y],end=' ' ) #avoid new line
            print("\n")
```

```python
a_rows = int(input("Enter A matrix's rows:"))
a_cols = int(input("Enter A matrix's cols:"))
print("Matrix A({}, {}):".format(a_rows, a_cols))
A = Matrix(a_rows, a_cols)
A.display()
    
b_rows = int(input("Enter B matrix's rows:"))
b_cols = int(input("Enter B matrix's cols:"))
print("Matrix B({}, {}):".format(b_rows, b_cols))
B = Matrix(b_rows, b_cols)
B.display()
    
print("="*10, 'A + B', "="*10)
result = A.add(B)
if result is not None:
    result.display()
    
print("="*10, 'A - B', "="*10)
result = A.sub(B)
if result is not None:
    result.display()
print("="*10, 'A * B', "="*10)
result = A.mul(B)
if result is not None:
    result.display()

print("="*5, "the transpose of A*B", "="*5)
if result is not None:
    result = result.transpose()
    
    result.display()
```
